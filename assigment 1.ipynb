Structured output:

```
from langchain_core.prompts import PromptTemplate

structured_prompt_template = """
Answer the following question and return the output in JSON format with keys:
- answer: final answer
- steps: list of reasoning steps

Question: {question}

JSON:
"""

structured_prompt = PromptTemplate(
    input_variables=["question"],
    template=structured_prompt_template
)

structured_chain = structured_prompt | llm

question = "If a train travels 60 km in 1 hour, how long will it take to travel 180 km?"

response = structured_chain.invoke({"question": question})

print(response)
```

Few Shot

```
few_shot_structured_template = """
Here are some examples (with JSON output):

Q: A car travels 100 km in 2 hours. How long will it take to travel 250 km?
A: {{
  "answer": "5 hours",
  "steps": [
    "Calculate speed: 100 km / 2 h = 50 km/h",
    "Calculate time: 250 km / 50 km/h = 5 hours"
  ]
}}

Q: A cyclist travels 15 km in 30 minutes. How far will they travel in 2 hours?
A: {{
  "answer": "60 km",
  "steps": [
    "Calculate speed: 15 km / 0.5 h = 30 km/h",
    "Calculate distance: 30 km/h * 2 h = 60 km"
  ]
}}

Now answer in the same JSON format:

Q: {question}
A:
"""
```

```
from langchain_core.prompts import PromptTemplate

few_shot_structured_prompt = PromptTemplate(
    input_variables=["question"],
    template=few_shot_structured_template
)

few_shot_structured_chain = few_shot_structured_prompt | llm

response = few_shot_structured_chain.invoke({
    "question": "If a train travels 60 km in 1 hour, how long will it take to travel 180 km?"
})

print(response)
```
